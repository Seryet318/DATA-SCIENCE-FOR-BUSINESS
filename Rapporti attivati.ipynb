### __Data cleaning and preparing__

From initial analysis, it emerged that there are ambiguities in the data, for example, there are people who are zero years old, underage or over a hundred years old. There are also ambiguities in the dates, for example dates that go beyond the date on which we find ourselves today.

In this session we will go to:

1. __Eliminate people of the wrong age.__
2. __Remove all null values present.__

___

1. __Eliminate people of the wrong age__

```python
dropAge = df_Rapporti_Attivati[(df_Rapporti_Attivati['ETA'] >= 70) | (df_Rapporti_Attivati['ETA'] <= 17)].index
```

```python
df_Rapporti_Attivati.drop(dropAge, inplace=True)
```

2. __Remove all null values present__

```python
col = ['DATA', 'GENERE', 'ETA', 'SETTOREECONOMICODETTAGLIO', 'TITOLOSTUDIO', 'CONTRATTO', 
       'MODALITALAVORO', 'PROVINCIAIMPRESA','ITALIANO']

for column in col:
    df_Rapporti_Attivati.loc[df_Rapporti_Attivati[column] == '?', column] = np.nan

df_Rapporti_Attivati.dropna(axis = 0, inplace = True) 
```

## __Data analysis before covid, during covid and Global__

After having carried out an exploratory analysis of the dataset and having cleaned the dataset from all the wrong data, we go to perform an analysis to see the progress of the reports activated:
1. __Performance of globally activated relationships.__
2. __Performance of pre-Covid activated reports.__
3. __Performance of the activated During-Covid relations.__
4. __Performance relations for the city of Varese.__

NOTE: during the analysis of the progress of the reports, we will consider the same time interva

```python
data=df_Rapporti_Attivati['DATA']
```

```python
year=[]
mounth=[]
day=[]
data_Complete=[]
for i in data:
    data1=datetime.strptime(i, "%d/%m/%Y")
    year.append(data1.year)
    mounth.append(data1.month)
    day.append(data1.day)
    data_Complete.append(data1)
```

```python
df_Rapporti_Attivati.drop(['DATA'], axis=1, inplace=True)
```

```python
df_Rapporti_Attivati['DATA'] = data_Complete
df_Rapporti_Attivati['YEAR'] = year
df_Rapporti_Attivati['MOUNTH'] = mounth
df_Rapporti_Attivati['DAY'] = day
```

1. __Global DataAnalysis__

```python
global_Data_YEAR = df_Rapporti_Attivati.groupby(['YEAR']).MOUNTH.count()
```

```python
global_Data_MOUNTH = df_Rapporti_Attivati.groupby(['MOUNTH']).YEAR.count()
```

```python
## crete a different subplot for two dataAnalysis
plt.style.use('seaborn-whitegrid')
figure = plt.figure(figsize=(10, 10))

x_dataY = global_Data_YEAR.index
y_dataY = global_Data_YEAR.values

x_dataM = global_Data_MOUNTH.index
y_dataM = global_Data_MOUNTH.values

plt.subplot(2, 1, 1)
plt.plot(x_dataY, y_dataY, color='orange')
plt.title('Distribution RAPPORTI_ATTIVATI for 1973 at 2021')
plt.xlabel('YEAR')
plt.ylabel('COUNT')

plt.subplot(2, 1, 2)
plt.plot(x_dataM, y_dataM, color='red')
plt.title('Distribution RAPPORTI_ATTIVATI for mounth (1973-2021)')
plt.xlabel('MOUNTH')
plt.ylabel('COUNT')
```
